# Model Analysis Notebook

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
npy_root_path = '../data/npy'
```

```python
y_pred_lin = np.load(f'{npy_root_path}/y_pred_lin.npy')
y_true_lin = np.load(f'{npy_root_path}/y_test_lin.npy')

y_pred_mlp = np.load(f'{npy_root_path}/y_pred_mlp.npy')
y_true_mlp = np.load(f'{npy_root_path}/y_test_mlp.npy')

y_pred_cnn = np.load(f'{npy_root_path}/y_pred_cnn.npy')
y_true_cnn = np.load(f'{npy_root_path}/y_test_cnn.npy')

y_pred = np.load(f'{npy_root_path}/y_pred.npy')
y_true = np.load(f'{npy_root_path}/y_test.npy')
```

linear model

```python
date_range = pd.date_range(start='2017-01-01', end='2018-12-31', periods=len(y_true))

plt.figure(figsize=(10, 6))
plt.plot(date_range, y_pred_lin, label='Predictions', color='red', linestyle='-')
plt.plot(date_range, y_true_lin, label='Ground Truth', color='blue')
plt.xlabel('time', fontsize=36)
plt.ylabel('wave height (m)', fontsize=36)
plt.title('linear regression', fontsize=36)
# plt.legend(fontsize=18)

plt.ylim([-5, 5])

plt.show()
```

analyzing mlp

```python

date_range = pd.date_range(start='2017-01-01', end='2018-12-31', periods=len(y_true))

plt.figure(figsize=(10, 6))
plt.plot(date_range, y_pred_mlp, label='Predictions', color='red', linestyle='-')
plt.plot(date_range, y_true_mlp, label='Ground Truth', color='blue')
plt.xlabel('time', fontsize=36)
plt.ylabel('wave height (m)', fontsize=36)
plt.title('multi-layer perceptron', fontsize=36)
plt.legend(fontsize=18)
plt.ylim([0, 5])

plt.show()
```

analyze CNN predictions against actual

```python
date_range = pd.date_range(start='2017-01-01', end='2018-12-31', periods=len(y_true))

plt.figure(figsize=(10, 6))
plt.plot(date_range, y_true_cnn, label='Ground Truth', color='blue')
plt.plot(date_range, y_pred_cnn, label='Predictions', color='red', linestyle='-')
plt.xlabel('time', fontsize=18)
plt.ylabel('wave height (m)', fontsize=18)
plt.title('convolutional neural network', fontsize=18)
plt.legend(fontsize=18)

plt.show()
```
